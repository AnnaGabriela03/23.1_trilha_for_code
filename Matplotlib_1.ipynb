```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x: float) -> float:
  return x**2 + 3
def f2(x: float) -> float:
  return x**3
```

```python
x = np.linspace(-100, 100, 1000)
y2 = f2(x)
y = f(x)
```

```python
# assim se plota o gráfico de linha
plt.plot(x, y)
plt.plot (x, y2)
plt.title("plotando a função x**2 + 3")
plt.xlabel("plotando o x")
plt.ylabel("plotanto o y")
plt.grid()
plt.show()
```

```python
# para plotar o gráfico de barra ou de pizza
plt.bar()
plt.pie()
```
